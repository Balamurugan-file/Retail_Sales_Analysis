```python
df.head()
```

```python
# Convert everything in the date column to the same format: YYYY-MM-DD

df['Date'] = pd.to_datetime(df['Date'],errors='coerce')
```

```python
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

```python
'''
"Data Cleaning: 126 records (approx. 9% of the dataset) were removed because they contained non-positive quantities (zeros or negatives).
These were treated as data entry errors or cancelled orders to ensure the analysis focuses on successfully completed sales."
'''

# Save the 'Clean' data back to the same variable
df = df[df['Quantity'] > 0]
```

```python
df.shape
```

```python
df['Price'].isna().sum()
```

```python
# Removing all non-numeric characters from Price column - Using regex pattern '[^\d.]'
# to strip currency labels like MMK, mmk, M.M.K, ks, kyats etc. keeping only digits and decimal point

# First, remove commas to handle large numbers like 2,000,000
df['Price'] = df['Price'].astype(str).str.replace(',', '')

# Extract ONLY the first number (handles integers or decimals correctly)
# This logic will ignore the messy dots in "M.M.K"
df['Price'] = df['Price'].str.extract(r'(\d+\.?\d*)')[0]

# Convert to numeric
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df['Price'].max()
```

```python
df['Price'].min()
```

```python
# Filling Discount_% null values based on (Product + Category) group median
# Using 'Median' over 'Mean' since discount levels are fixed (0, 5, 10, 15)
# and median always returns an actual existing discount value unlike mean which gives decimals

df['Discount_%'] = df['Discount_%'].fillna(
    df.groupby(['Product','Category'])['Discount_%'].transform('median'))
```

```python
# Total_Sales column - No action required
# Rename the column name for better understanding and analysis
df = df.rename(columns={'Total_Sales': 'Gross_Sales_(USD)'})
# Column is no null values after removing invalid Quantity rows
```

```python
df.isna().sum()
```

```python
# Standardising Region column - fixing case, spaces and abbreviations
df['Region'] = df['Region'].str.strip().str.title()
```

```python
# Mapping only confirmed abbreviations and wrong spellings to correct standard region names
# Tanintharyi and Magway are kept as-is - they are real valid Myanmar regions

df['Region'] = df['Region'].replace({
    'Mdy': 'Mandalay',
    'Ygn': 'Yangon',
    'Npd': 'Naypyidaw',
    'Nay Pyi Daw': 'Naypyidaw'
})
```

```python
# Filling Region null values using Mode - correct imputation for categorical columns

df['Region'] = df['Region'].fillna(df['Region'].mode()[0])
```

```python
df['Region'].unique()
```

```python
# Filling Sales_Channel null values using Mode - correct imputation for categorical columns

df['Sales_Channel'] = df['Sales_Channel'].str.strip().str.title()

df['Sales_Channel'] = df['Sales_Channel'].fillna(df['Sales_Channel'].mode()[0])
```

```python
df['Sales_Channel'].unique()
```

```python
# Filling Payment_Method null values using Mode - correct imputation for categorical columns

df['Payment_Method'] = df['Payment_Method'].str.strip().str.title()

df['Payment_Method'] = df['Payment_Method'].fillna(df['Payment_Method'].mode()[0])
```

```python
df['Payment_Method'].unique()
```

```python
# Filling Customer_Type null values using Mode - correct imputation for categorical columns

df['Customer_Type'] = df['Customer_Type'].str.strip().str.title()

df['Customer_Type'] = df['Customer_Type'].fillna(df['Customer_Type'].mode()[0])
```

```python
df['Customer_Type'].unique()
```

```python
# Filling Customer_Rating null values using Median - preferred over Mode for numeric rating scales (1-5)
# Median represents the true middle value and is not affected by frequency dominance

df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())
```

```python
df['Customer_Rating'].unique()
```

```python
# Step 1 - Replacing invalid -1 values with NaN in Delivery_Days
# -1 delivery days is impossible - treating as missing data entry error

import numpy as np

df['Delivery_Days'] = df['Delivery_Days'].replace(-1,np.nan)

# Step 2 - Filling all Delivery_Days null values using Median
# Median is correct for numeric ordered values - not affected by extreme values

df['Delivery_Days'] = df['Delivery_Days'].fillna(df['Delivery_Days'].median())
```

```python
df['Delivery_Days'].unique()
```

```python
df.isna().sum()
```

```python
# Calculate Gross Sales in MMK (Price * Quantity)
df['Gross_Sales_MMK'] = df['Price'] * df['Quantity']
```

```python
# Calculate Net Sales in MMK (Apply the discount)
df['Net_Sales_MMK'] = df['Gross_Sales_MMK'] * (1 - df['Discount_%'] / 100)
```

```python
# Calculate Net Sales in USD (Apply the discount)
df['Net_Sales_USD'] = df['Net_Sales_MMK'] / 2500

# Round to 2 decimal places for a clean look
df['Net_Sales_USD'] = df['Net_Sales_USD'].round(2)
```

```python
df.head()
```

```python
df.isna().sum()
```

#**Treating Outliers**

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns   #getting only numerical columns for finding outliers
numerical_columns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#identifying outliers using Boxplots

plt.figure(figsize=(13,10))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 4, i)   # 4 rows, 4 columns layout
    sns.boxplot(x=df[col])
    plt.title(f'Distribution of {col}')

plt.suptitle('Boxplots of Numerical Columns', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
#identifying outliers using KDEplots

plt.figure(figsize=(13,10))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 4, i)   # 4 rows, 4 columns layout
    sns.kdeplot(x=df[col])
    plt.title(f'Distribution of {col}')

plt.suptitle('KDE Plots of Numerical Columns', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python

# Price column - No outlier treatment needed
# The single dot seen in the boxplot is NOT a real outlier — it is the Laptop price of 2,000,000 MMK which is a valid real price point in this dataset.
# 95th percentile check returned 0 rows above threshold - no action required

pr_threshold = df['Price'].quantile(0.95)

print('95th percentile:',pr_threshold)

df[df['Price'] > pr_threshold].shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Price'])
plt.show()
```

```python
# Treating outlier using IQR method
# Outliers are retained as they represent real business transactions

# Step 1: Calculate IQR
Q1 = df['Gross_Sales_(USD)'].quantile(0.25)
Q3 = df['Gross_Sales_(USD)'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define bounds
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Step 3: Count outliers
outliers = df[(df['Gross_Sales_(USD)'] < lower_bound) | (df['Gross_Sales_(USD)'] > upper_bound)]
print("Number of Outliers:", outliers.shape[0])

# Step 4: Display top outliers
print(outliers[['Product', 'Gross_Sales_(USD)']].sort_values(by='Gross_Sales_(USD)', ascending=False).head())
```

```python
lower = df['Gross_Sales_(USD)'].quantile(0.01)
upper = df['Gross_Sales_(USD)'].quantile(0.99)

df['Gross_Sales_(USD)_clipped'] = df['Gross_Sales_(USD)'].clip(lower, upper)
```

```python
# Boxplot after handling outliers

sns.boxplot(x=df['Gross_Sales_(USD)_clipped'])
plt.title("Outlier Detection - Gross_Sales_(USD)")
plt.show()
```

**Note:Outliers in the Gross and Net Sales columns were detected using the IQR method. Although several outliers were identified, they were retained because they represent
real-world high-value sales transactions such as bulk purchases.  Removing or capping these values could distort business insights.**
